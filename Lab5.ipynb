## **1.2: Introduction**
---

Least Squares Regression is one of the most fundamental techniques in data science, statistics, and machine learning. It is used to find the best-fitting line (or curve) that minimizes the sum of squared differences between observed values and predicted values. This method is essential because real-world data is often noisy, and we need a systematic way to model relationships between variables.

<a name="p3"></a>

---
## **1.3: Definition of Least Squares Regression Line**
---

Least Squares Regression is a statistical method used to determine the equation of the line that best fits a given set of data points by minimizing the sum of squared residuals (the differences between observed and predicted values). The equation of a simple linear regression line is:


$\hat{y} = \beta_0 + \beta_1\cdot x$


where:
- \( $y$ \) is the dependent variable (response),
- \( x \) is the independent variable (predictor),
- \( $\beta_1$ \) is the slope (rate of change),
- \( $\beta_0$ \) is the y-intercept.

### **Equation of the Least-Squares Regression Line using Calculus**

The equation of the least-squares regression line is given by:


$\hat{y} = \beta_1x + \beta_0$

where:


$\beta_1 = \frac{S_{xy}}{S_{x^2}}$


is the **slope** of the least-squares regression line, or:


$\beta_1 = \frac{n \sum_{i=1}^{n} x_i y_i - \sum_{i=1}^{n} x_i \sum_{i=1}^{n} y_i}{n \sum_{i=1}^{n} x_i^2 - \left(\sum_{i=1}^{n} x_i\right)^2}$



$\beta_0 = \bar{y} - b_1 \bar{x}$

is the **y-intercept** of the least-squares regression line, or:


$\beta_0 = \frac{1}{n}\left(\sum_{i=1}^{n} y_i - b_1 \sum_{i=1}^{n} x_i\right)$

## **Example: Student Absences and Final Grades**
Mr. Pineda is teaching a data science class and wants to analyze the relationship between the number of absences and students' final grades. He collects the following data:

| Student | Absences ( x ) | Final Grade ( y ) |
|---------|----------------|----------------|
| 1       | 2              | 96             |
| 2       | 4              | 92             |
| 3       | 6              | 88             |


By applying the Least Squares Regression method, he can determine the best-fit line equation that models how absences affect final grades. This allows him to predict a student's final grade based on their number of absences.

## **Why is this Important?**
- **Predictive Analysis:** It helps forecast outcomes based on trends in data.
- **Decision-Making:** Businesses and educators use regression models to make informed decisions.
- **Identifying Relationships:** It helps establish the strength and direction of relationships between variables.
- **Error Minimization:** The least squares method ensures the best possible fit by reducing errors.

By mastering Least Squares Regression, students gain essential skills for analyzing and interpreting real-world data, making them valuable in various fields such as economics, engineering, and machine learning.

## **Conclusion**
Least Squares Regression is a powerful tool in data analysis, helping researchers and professionals make data-driven decisions. In the case of Mr. Pineda's class, understanding the effect of absences on final grades provides insights that can improve student performance and academic strategies.


```python
# prompt: Can you please take the data below and do a least square regression analysis?

import numpy as np

# Sample data (replace with your actual data)
x = np.array([2, 4, 6])
y = np.array([96, 92, 88])

# Calculate the mean of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the slope (beta_1)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
beta_1 = numerator / denominator

# Calculate the y-intercept (beta_0)
beta_0 = y_mean - beta_1 * x_mean

# Print the results
print("Slope (beta_1):", beta_1)
print("Y-intercept (beta_0):", beta_0)

# You can now use the equation y_hat = beta_0 + beta_1 * x to predict y values
# for new x values.
```

```python


import numpy as np


# **Matrix Least Sqare Regression**

X = np.array([[1, 2],[1, 4], [1,6]])
Y = np.array([[96],[92],[88]])  # Y should be a column vector

XT = np.transpose(X)
XTX = np.dot(XT,X)
# use pinv instead of inv
beta_hat = np.linalg.pinv(XTX) @ XT @ Y

print("\nbeta_hat (Matrix Least Squares):")
print(beta_hat)

# Extract beta_0 and beta_1 from beta_hat
beta_0_matrix = beta_hat[0][0]
beta_1_matrix = beta_hat[1][0]


# Example prediction using the matrix method
print("y_hat =", beta_0_matrix, '+', beta_1_matrix, '*x' )
```

# **Least Squares Regression: Understanding the Impact of Absences and Quiz Scores on Final Grades**

## **Introduction**  
Mr. Pineda is teaching a data science class and wants to analyze the relationship between students' final grades, their number of absences, and their quiz scores. He collects the following data:  

| Student | Absences ( $x_1$ ) | Quiz Score ( $x_2$ ) | Final Grade ( y ) |
|---------|----------------|----------------|----------------|
| 1       | 2              | 80             | 96             |
| 2       | 4              | 85             | 92             |
| 3       | 6              | 90             | 88             |

Using **Least Squares Regression** for multiple variables, he can determine a model to predict a student's final grade based on their number of absences and quiz scores.

---

## **Finding the Multiple Regression Equation**  

A multiple regression model follows this form:


$\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2$


where:
- \( $y$ \) is the final grade,
- \( $x_1$ \) is the number of absences,
- \( $x_2$ \) is the quiz score,
- \( $\beta_0$ \) is the intercept,
- \( $\beta_1$ \) is the coefficient for absences,
- \( $\beta_2$ \) is the coefficient for quiz scores.

---

## **Conclusion**  
This regression model helps Mr. Pineda understand the impact of **both absences and quiz performance** on final grades. With this, he can predict future student performance and advise students accordingly.


```python
import numpy as np

X = np.array([[1, 2, 80],[1, 4, 85], [1,6,90]])
Y = np.array([[96],[92],[88]])  # Y should be a column vector

XT = np.transpose(X)
XTX = np.dot(XT,X)
# use pinv instead of inv
beta_hat = np.linalg.pinv(XTX) @ XT @ Y

beta_hat
```

```python
import numpy as np


# **Matrix Least Sqare Regression**

X = np.array([[1, 2, 80],[1, 4, 85], [1,6,90]])
Y = np.array([[96],[92],[88]])  # Y should be a column vector

XT = np.transpose(X)
XTX = np.dot(XT,X)
# use pinv instead of inv
beta_hat = np.linalg.pinv(XTX) @ XT @ Y

print("\nbeta_hat (Matrix Least Squares):")
print(beta_hat)

# Extract beta_0 and beta_1 from beta_hat
beta_0_matrix = beta_hat[0][0]
beta_1_matrix = beta_hat[1][0]
beta_2_matrix = beta_hat[2][0]


# Example prediction using the matrix method
print("y_hat =", beta_0_matrix, '+', beta_1_matrix, '*x1',"+", beta_2_matrix, '*x2' )
```

```python
# prompt: Can you please graph the data in 3D and insert the least square regression?| Student | Absences ( $x_1$ ) | Quiz Score ( $x_2$ ) | Final Grade ( y ) |
# |---------|----------------|----------------|----------------|
# | 1       | 2              | 80             | 96             |
# | 2       | 4              | 85             | 92             |
# | 3       | 6              | 90             | 88             |

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x1 = np.array([2, 4, 6])
x2 = np.array([80, 85, 90])
y = np.array([96, 92, 88])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(x1, x2, y, c='r', marker='o')

# Calculate the least squares regression plane
X = np.column_stack((np.ones_like(x1), x1, x2))
beta_hat = np.linalg.lstsq(X, y, rcond=None)[0]

# Create a meshgrid for the plane
x1_plane = np.linspace(min(x1), max(x1), 10)
x2_plane = np.linspace(min(x2), max(x2), 10)
X1_plane, X2_plane = np.meshgrid(x1_plane, x2_plane)
Y_plane = beta_hat[0] + beta_hat[1] * X1_plane + beta_hat[2] * X2_plane

# Plot the regression plane
ax.plot_surface(X1_plane, X2_plane, Y_plane, alpha=0.5)

# Set axis labels
ax.set_xlabel('Absences (x1)')
ax.set_ylabel('Quiz Score (x2)')
ax.set_zlabel('Final Grade (y)')

# Show the plot
plt.show()
```

# **Least Squares Regression: Understanding the Impact of Absences, Quiz Scores, and Exam Scores on Final Grades**

## **Introduction**  
Mr. Pineda is teaching a data science class and wants to analyze the relationship between students' final grades, their number of absences, quiz scores, and exam scores. He collects the following data:  

| Student | Absences ($x_1$) | Quiz Score ($x_2$) | Exam 1 Score ($x_3$) | Final Grade ( y) |
|---------|----------------|----------------|----------------|----------------|
| 1       | 2              | 80             | 85             | 96             |
| 2       | 4              | 85             | 88             | 92             |
| 3       | 6              | 90             | 92             | 88             |

Using **Least Squares Regression** for multiple variables, he can determine a model to predict a student's final grade based on their number of absences, quiz scores, and exam scores.

---

## **Finding the Multiple Regression Equation**  

A multiple regression model follows this form:


$\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_3$


where:
- \( y \) is the final grade,
- \( $x_1$\) is the number of absences,
- \( $x_2$\) is the quiz score,
- \( $x_3$\) is the exam score,
- \( $\beta_0$ \) is the intercept,
- \( $\beta_1$ \) is the coefficient for absences,
- \( $\beta_2$  \) is the coefficient for quiz scores,
- \( $\beta_3$ \) is the coefficient for exam scores.

---

# **Matrix Form for Linear Regression**

## **Introduction**
Least Squares Regression can be represented in **matrix form**, which provides a more compact and efficient way to solve for regression coefficients. In this section, we will introduce the **matrix representation** for multiple linear regression and apply it to our dataset.

---

## **General Matrix Form of Linear Regression**
A multiple linear regression model follows the equation:


$\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n$


This can be rewritten in **matrix form** as:


$Y = X \beta + \varepsilon$

where:
- \( Y \) is the **vector of observed values** (final grades in our case),
- \( X \) is the **design matrix** (containing ones for the intercept and values for each predictor),
- \( $\beta$ \) is the **vector of coefficients** (intercept and regression coefficients),
- \( $\varepsilon$ \) is the **error vector** (difference between observed and predicted values).

### **Expanded Matrix Representation**
For our dataset:

| Student | Absences ( $x_1$ ) | Quiz Score ($x_2$) | Exam 1 Score ($x_3$) | Final Grade (y) |
|---------|----------------|----------------|----------------|----------------|
| 1       | 2              | 80             | 85             | 96             |
| 2       | 4              | 85             | 88             | 92             |
| 3       | 6              | 90             | 92             | 88             |

We express this in **matrix form**:


$Y =
\begin{bmatrix}
96 \\
92 \\
88
\end{bmatrix},$

$X =
\begin{bmatrix}
1 & 2 & 80 & 85 \\
1 & 4 & 85 & 88 \\
1 & 6 & 90 & 92
\end{bmatrix}, \quad
\beta =
\begin{bmatrix}
\beta_0 \\
\beta_1 \\
\beta_2 \\
\beta_3
\end{bmatrix}$


The least squares estimate for \( $\beta$ \) is given by:


$\hat{\beta} = (X^T X)^{-1} X^T Y$


This equation provides the best-fit coefficients that minimize the sum of squared errors, giving us a predictive model for final grades based on absences, quiz scores, and exam scores.

```python
import numpy as np


# **Matrix Least Sqare Regression**

X = np.array([[1, 2, 80, 85],[1, 4, 85, 88], [1,6,90,92]])
Y = np.array([[96],[92],[88]])  # Y should be a column vector

XT = np.transpose(X)
XTX = np.dot(XT,X)
# use pinv instead of inv
beta_hat = np.linalg.pinv(XTX) @ XT @ Y

print("\nbeta_hat (Matrix Least Squares):")
print(beta_hat)

# Extract beta_0 and beta_1 from beta_hat
beta_0_matrix = beta_hat[0][0]
beta_1_matrix = beta_hat[1][0]
beta_2_matrix = beta_hat[2][0]
beta_3_matrix = beta_hat[3][0]

# Example prediction using the matrix method
print("y_hat =", beta_0_matrix, '+', beta_1_matrix, '*x1',"+", beta_2_matrix, '*x2', beta_3_matrix, '*x3' )
```
