## **1.2: Introduction**
---

In the study of algebra and systems of equations, a system of linear equations consists of multiple linear equations that are solved simultaneously. So far, we've explored the fundamental possibilities for solutions to such systems: **no solution**, **a unique solution**, and **infinitely many solutions**. These outcomes arise based on the relationship between the equations and how they intersect, either at a point, not at all, or along a line (or plane in higher dimensions).

However, the utility of systems of linear equations extends far beyond simply determining whether a solution exists. One key application is partial fraction decomposition, where systems of linear equations are used to express a rational function as a sum of simpler fractions, which is crucial for integral calculus, particularly in the context of solving integrals involving rational functions. This involves breaking down complex expressions into components that can be more easily integrated.

Additionally, systems of linear equations play a crucial role in solving nonlinear systems of equations. While linear systems deal with equations of straight lines, nonlinear systems can involve more complex relationships, such as curves or surfaces. Techniques for solving linear systems are often adapted to help solve nonlinear systems, bridging the gap between algebraic methods and more advanced mathematical modeling in fields such as physics, economics, and engineering.

In this exploration, we will examine not only how to solve linear systems, but also how these solutions can be applied in various mathematical and real-world contexts, including those involving partial fractions and nonlinear systems.

## **1.3: Partial Fraction Demoposition**
---

**Example:**  $\frac{3}{x+2} + \frac{7}{(x-4)}$

We want a method of reversing this process.

**Note:** We only do the partial fraction decomposition on **proper** ratial expressions (the degree of the numerator is less than the degree of the denominator).

# Partial Fraction Decomposition & Integration

Given:

$\frac{10x + 2}{(x - 4)(x + 2)}$


We decompose:

$\frac{10x + 2}{(x - 4)(x + 2)} = \frac{A}{x - 4} + \frac{B}{x + 2}$


Multiplying by $(x - 4)(x + 2)$:


$10x + 2 = A(x + 2) + B(x - 4)$


Solving for \( A \) and \( B \):

$A = 7, \quad B = 3$

So,


$\frac{10x + 2}{(x - 4)(x + 2)} = \frac{7}{x - 4} + \frac{3}{x + 2}$

### Integral:


$\int \frac{10x + 2}{(x - 4)(x + 2)} \,dx = \int \left( \frac{7}{x - 4} + \frac{3}{x + 2} \right) \,dx
$
$= 7 \ln |x - 4| + 3 \ln |x + 2| + C$




```python
import sympy as sp

x = sp.symbols('x')

function = ((7/(x-4)) + (3/(x+2)))
result = sp.integrate(function, x)

print(result)
```

# Partial Fraction Decomposition Steps

| Factor in denominator | Term(s) in decomposition |
|----------------------------|--------------------------------------------|

3.   Multiply both sides of the resulting equation by the least common denominator, which is the orginal denominator. Use the distributive property, combine like terms, and factor out each different power of *x*. Next equate coefficients.

Practice 1: Write the partial fraction decomposition for each rational expression.

$\frac{10x+2}{(x+2)(x-4)}$=



```python
from sympy import symbols, Eq, solve

x = symbols('x')
A, B = symbols('A B')

#Define the equation
expression = (10*x + 2) / ((x + 2) * (x - 4))

#Partial fraction decomposition setup
decomposition = A/(x + 2) + B/(x - 4)

#Find the common denominator by multiplying both sides by the LCD
equation = Eq((10*x+2), A*(x-4) + B*(x+2))

#Solve for A and B
solution = solve((equation), (A, B))

#Print the solution
print(f"Solution is: {solution}")
print(f"The partial fraction decomposition of {expression} is:")
print(f'{solution[A]}/(x+2), "+", {solution[B]}/(x-4)')
```

Exercise 1:  Write the partial fraction decomposition for each rational expression.

$\frac{x-3}{(x+2)(x+1)^2}$=

```python
# prompt: Can you please solve the fraction decomposition in python for x3(x+2)(x+1)2 =

import sympy as sp
from sympy import *
from sympy import symbols, Eq, solve

x = symbols('x')
A, B, C = symbols('A B C')

# Define the equation
expression = (x - 3) / ((x + 2) * (x + 1)**2)

# Partial fraction decomposition setup
decomposition = A/(x + 2) + B/(x + 1) + C/((x + 1)**2)

# Find the common denominator by multiplying both sides by the LCD
equation = Eq((x - 3), A*(x + 1)**2 + B*(x + 2)*(x + 1) + C*(x + 2))

# Solve for A, B, and C
solution = solve(equation, (A, B, C))

# Print the solution
print(f"Solution is: {solution}")
print(f"The partial fraction decomposition of {expression} is:")
print(f'{solution[A]}/(x+2) + {solution[B]}/(x+1) + {solution[C]}/((x+1)**2)')
```

Practice 2: Find real numbers a, b, and c so that the graph of the fuction $y=ax^2+bx+c$ contains the points (-1,3), (3,8), and (0,2).


```python
# prompt: Can you please help me write the code in python to solve:  Find real numbers a, b, and c so that the graph of the fuction  y=ax2+bx+c  contains the points (-1,3), (3,8), and (0,2).

import sympy as sp
from sympy import *
from sympy import symbols, Eq, solve

a, b, c = symbols('a b c')

# Define the equations based on the given points
eq1 = Eq(a*(-1)**2 + b*(-1) + c, 3)  # (-1, 3)
eq2 = Eq(a*(3)**2 + b*(3) + c, 8)    # (3, 8)
eq3 = Eq(a*(0)**2 + b*(0) + c, 2)    # (0, 2)

# Solve the system of equations
solution = solve((eq1, eq2, eq3), (a, b, c))

# Print the solution
print(f"Solution is: {solution}")
print(f"a = {solution[a]}, b = {solution[b]}, c = {solution[c]}")
```

Exercise 3: Solve each system of equations.


$log_x(2y)=3$

$log_x(4y)=2$

```python
# prompt: Can you please help me solve this in Python?:
# Solve each system of equations.
# logx(2y)=3
# logx(4y)=2

import sympy as sp
from sympy import *

x, y = symbols('x y', positive=True)  # Specify x and y as positive

# Define the equations
eq1 = Eq(log(2*y, x), 3)
eq2 = Eq(log(4*y, x), 2)

# Solve the system of equations
solution = solve((eq1, eq2), (x, y))

# Print the solutions
print(f"Solutions: {solution}")

# Further processing solutions
for sol in solution:
    x_val, y_val = sol
    print(f"x = {x_val}, y = {y_val}")
    print(f"Verification:")
    print(f"log({2*y_val}, {x_val}) = {log(2*y_val,x_val)}")
    print(f"log({4*y_val}, {x_val}) = {log(4*y_val,x_val)}")

```

```python
# prompt: Can you please solve each system of equations in a graph:
# logx(2y)=3
# logx(4y)=2

import sympy as sp
from sympy import *
from sympy import symbols, Eq, solve
import plotly.graph_objects as go
import numpy as np

x, y = symbols('x y', positive=True)  # Specify x and y as positive

# Define the equations
eq1 = Eq(log(2*y, x), 3)
eq2 = Eq(log(4*y, x), 2)

# Solve the system of equations
solution = solve((eq1, eq2), (x, y))

# Print the solutions
print(f"Solutions: {solution}")

# Further processing solutions
for sol in solution:
    x_val, y_val = sol
    print(f"x = {x_val}, y = {y_val}")
    print(f"Verification:")
    print(f"log({2*y_val}, {x_val}) = {log(2*y_val,x_val)}")
    print(f"log({4*y_val}, {x_val}) = {log(4*y_val,x_val)}")

#For graphing, we need to convert the equations to a form suitable for plotting.
#log_x(2y)=3 => x^3 = 2y => y = x^3/2
#log_x(4y)=2 => x^2 = 4y => y = x^2/4

x_vals = np.linspace(0.1, 10, 400) #avoid x=0 and use a suitable range.

y_vals_eq1 = x_vals**3 / 2
y_vals_eq2 = x_vals**2 / 4

# Create the Plotly figure
fig = go.Figure()

# Add the curves
fig.add_trace(go.Scatter(x=x_vals, y=y_vals_eq1, mode='lines', name='log_x(2y)=3'))
fig.add_trace(go.Scatter(x=x_vals, y=y_vals_eq2, mode='lines', name='log_x(4y)=2'))

# Set the title and axis labels
fig.update_layout(title='Graph of the log_x(2y)=3 and log_x(4y)=2',
                  xaxis_title='x',
                  yaxis_title='y',
                  legend_title = 'Legend')


# Show the plot
fig.show()
```
